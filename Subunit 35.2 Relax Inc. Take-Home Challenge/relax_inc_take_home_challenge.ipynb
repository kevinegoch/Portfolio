# 2 Feature Engineering

```python
df.tail()
```

```python
# Converting column of interest to a datetime object
df['time_stamp'] = pd.to_datetime(df['time_stamp']) - pd.to_timedelta(7, unit='d')

# Deriving a 7-day-period (week) from the time_stamp column
df['wk'] = df['time_stamp'].dt.isocalendar().week

# Grouping users by frequency of logins in a week
group = df.groupby(['wk','user_id'])['time_stamp'].count().reset_index(name='freq')

# Creating new column to indicate whether the user is adopted
# Note, an "adopted  user" is defined, according to the problem statement, 
# as  a  user  who   has  logged  into  the  product  on  three  separate
# days  in  at  least  one 7-day-period.
df['adopted'] = group.freq >= 3
df['adopted'].value_counts()
```

```python
df.groupby(['wk', 'user_id'])['adopted'].count().reset_index(name='freq')
```

```python
df.info()
```

## 2.1 Setting Up for Classification

```python
# Dropping some variables
dropped = ['name', 'email', 'visited', 'wk', 'creation_time', 'time_stamp']

# Implementing the dropping
df.drop(dropped, axis=1, inplace=True)
```

```python
# Binarizing adopted column
df['adopted'] = df['adopted'].astype(int)

# Verifying
df.adopted.value_counts()
```

```python
# One-hot encoding nominal object
df = pd.get_dummies(df, columns=['creation_source'])

# Verifying
df.info()
```

## 2.2 Splitting Data

```python
# Importing train_test_split
from sklearn.model_selection import train_test_split

# Getting predictor variables and response variable
X = df.drop('adopted', axis=1)
y = df['adopted'] # Getting reponse variable

# Shape of variables
print(X.shape, y.shape)

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# 3 Modeling

## 3.1 Random Forest

```python
# Importing essential modules.
from sklearn.ensemble        import RandomForestClassifier
from sklearn                 import preprocessing
from sklearn.metrics         import accuracy_score
from sklearn.metrics         import classification_report
from sklearn.metrics         import confusion_matrix
```

```python
# Instantiate, fit, and predict with RandomForestClassifier model
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Checking performance of DecisionTreeClassifier model
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## 3.2 Feature Importance

```python
# Checking most important features again
importance = model.feature_importances_

# Summarizing feature importance
for i, v in enumerate(importance):
    print('Feature: {}, Score: {}'.format(i, v))
    
# Plotting feature importance
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

It appears that features 0, 3, and 4 mostly predict user adoption. (See report for further details.)
